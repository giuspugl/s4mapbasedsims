```python
import healpy as hp 
import pylab as pl 
import glob 

```

```python
debug=True 
```

# Foreground  maps 

```python
for file in glob.glob("/global/cscratch1/sd/zonca/cmbs4/s4refsimtool/s4_reference_design_foregrounds/*") : 
    tele =file.split('/')[-1].split('-')[0]
    label=file.split('/')[-1] 
    if tele=='LAT' and debug :
        print('skipping LAT for debug ')
        continue 
    #pl.figure(figsize=(15,5)) 
    
    m =hp.read_map(glob.glob(file+'/*')[0] , field=[0,1,2], verbose=False)
    #hp.gnomview(m[0],rot=[40,-55], sub=131, title=label , xsize=400, ysize=200, reso=10 , min=-1e-5, max=1e-5)
    #hp.gnomview(m[1],rot=[40,-55], sub=132, title=label , xsize=400, ysize=200, reso=10 , min=-1e-6, max=1e-6)
    #hp.gnomview(m[2],rot=[40,-55], sub=133, title=label , xsize=400, ysize=200, reso=10 , min=-1e-6, max=1e-6)
    cl=hp.anafast(m, lmax=1000 )
    pl.figure() 
    
    pl.title(f" { file.split('/')[-2]} {label}" )
    ell = pl.arange(cl.shape[1] )
    #pl.subplot(131)
    pl.semilogy(ell,ell**2* cl[0]*1e12,  label='TT ' )
    pl.xlim([40,ell.max() ])
    pl.semilogy(ell,ell**2* cl[1]*1e12,label='EE'  )
    #pl.subplot(133)
    pl.semilogy(ell,ell**2* cl[2]*1e12, label='BB') 
    pl.ylabel(r'$D_{\ell} \, [uK^2]  $', fontsize=15)
    pl.legend()
    #break 
    #print (glob.glob(file+'/*') ) 
    
```

- weird morphology of spectra for HFS (wiggles at $\ell\sim 300$ ), ?
- hint of a bump in TT  at small angular scales for MFLS and MFHS 
- no EE/BB asymmetry ( E-to-B leakage of cut sky )
